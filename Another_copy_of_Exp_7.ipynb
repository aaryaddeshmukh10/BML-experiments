```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/iris/iris.csv'

column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

df = pd.read_csv(file_path, header = None, names = column_names)

print(df.head())

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
df

x = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df['class']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print("Training size:", x_train.shape)
print("Testing size:", x_test.shape)
```

*Manual Implementation*

```python
from collections import Counter
import numpy as np

# Manual KNN function
def knn_predict(x_train, y_train, X_test, k):
    predictions = []

    for test_sample in X_test.values:
        # Compute Euclidean distances
        distances = np.sqrt(np.sum((x_train.values - test_sample) ** 2, axis=1))

        # Get indices of k nearest neighbors
        k_neighbors = np.argsort(distances)[:k]

        # Get labels of k nearest neighbors
        k_labels = [y_train.iloc[i] for i in k_neighbors]

        # Find most common class label
        most_common = Counter(k_labels).most_common(1)[0][0]
        predictions.append(most_common)

    return np.array(predictions)

# Set k value
k = 5

# Predict using manual KNN
y_pred_manual = knn_predict(x_train, y_train, x_test, k)

# Calculate accuracy
accuracy_manual = np.mean(y_pred_manual == y_test.values)

print(f"Manual KNN Accuracy with k={k}: {accuracy_manual:.2%}")
```

*Built-in method*

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize KNN model
model = KNeighborsClassifier(n_neighbors=20)

# Fit the model
model.fit(x_train, y_train)

# Predictions
y_pred = model.predict(x_test)

# Evaluation
print("Testing Score:", model.score(x_test, y_test))
print("Training Score:", model.score(x_train, y_train))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
